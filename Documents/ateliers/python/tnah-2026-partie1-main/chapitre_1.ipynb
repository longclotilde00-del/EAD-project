🙑 Eugène Atget (1857‑1927) est un photographe français, pionnier de la photographie documentaire, actif au tournant du XX<sup>e</sup> siècle. Il a construit une œuvre photographique monumentale qui documente avec rigueur le Paris de la Belle Époque, de la Zone et ses environs, avant et pendant les profondes transformations urbaines du début du XXᵉ siècle. Ses images décrivent les rues, les façades, les boutiques, les cours, les jardins et les scènes de la vie quotidienne avec une précision visuelle remarquable, cherchant moins à créer une esthétique artificielle qu’à enregistrer minutieusement la réalité urbaine. Atget n’aspirait pas à être vu comme un artiste au sens traditionnel : il envisageait la photographie d’abord comme un outil documentaire permettant de fixer des éléments de la ville menacés de disparition. Cependant, au fil du temps Atget évolue de la simple documentation architecturale (1890-1905) et adopte un regard plus attentif aux traces de vie humaine et sociale, photographiant artisans, travailleurs des rues et habitants de la zone. L’accumulation de ces vues confère à son œuvre une puissance métaphysique et poétique qui a séduit bien au‑delà de son intention initiale, influençant notamment les surréalistes et des générations de photographes modernes. Dans sa diversité, l’œuvre d’Atget se situe à l’intersection du document social et historique et d’une forme naissante de lecture critique de l’espace urbain, interrogeant la visibilité des corps et des constructions dans le paysage citadin.
Aujourd’hui, la BnF conserve plus de 4 500 épreuves positives d’Atget, réparties en séries thématiques élaborées par lui‑même ou par la bibliothèque, ainsi que plusieurs albums qu’il a constitués. Cette immense archive, complétée par d’autres fonds publics et privés, fait d’Atget une figure centrale de la photographie documentaire et un témoignage unique du vieux Paris, oscillant entre la constatation archivistique et l’évocation sensible du temps qui passe. 🙗

## 🙤 Chapitre 1 : à la découverte des métadonnées structurées de la BnF
### Gallica.bnf.fr : le portail des numérisations
Près de 4 000 de ses clichés ont été numérisés par la BnF et sont disponibles sur Gallica, organisés en quatre groupes : *petits métiers des rues*, *vieilles boutiques*, *habitants de la "zone"* et *éléments décoratifs*.  On peut entrer dans le corpus par l'une de ces catégories à partir de la page de présentation dédiée : https://gallica.bnf.fr/conseils/content/eugene-atget.

💡 Si vous le souhaitez, n'hésitez pas à prendre quelques minutes pour explorer le corpus au gré des rues de Paris il y a presque 130 ans !

> 🎬 Entrez dans le groupe `Vieilles Boutiques` depuis la page de présentation du corpus puis cliquez sur le titre de la première ressource intitulée `
Cabaret du Soleil d'or : R. St Sauveur : 7.bre [septembre] 1899 : [photographie] / [Atget] 1899` pour arriver sur une page affichant les numérisations de trois tirages de la même photographie. 
> Cliquer sur le lien `🛈 Informations détaillées` de la première ressource numérisée affiche sa notice, qui donne son identifiant unique ARK ainsi qu'un ensemble de métadonnées descriptives. Ce sont typiquement ces métadonnées que l'on souhaite récupérer pour alimenter notre base de connaissance.
> Profitez-en pour relever dès maintenant dans la cellule suivante 🏗️ les champs qui fournissent :
> 1. des indices de l**ocalisation géographique**
> 2. une **information temporelle**
> Notez également son **identifiant ARK**, qui identifie cette numéisation de manière univoque parmi toutes les ressources de la BnF. Nous en aurons aurons besoin plus tard.


**🏗️  - Cabaret du Soleil d'or : R. St Sauveur [...], numérisation n°1**

Champs de métadonnées donnant des indices géographiques ou temporels : 
- ...
Identifiant ARK** : ...


Cette notice contient un champ `Notice d'ensemble` renvoyant à une ressource du **catalogue général** de la BnF, elle aussi identifiée par un ARK.
Cet URL est construit à l'aide de l'ARK de la ressource. Ce type d'URL, qui identifie de manière **unique** et **persitente** une ressource sur le Web est appelée un URI (Uniform Resource Identifier) :
- Unique, car cet identifiant désigne une seule et unique ressource et la BnF garantit qu'aucune autre ressource ne porte le même identifiant.
- Persistante, car la BnF garantit que cet URL ne "cassera" pas, peu importe l'évolution technique des outils & portails de la bibliothèque. Même si les sites de la BnF changent, l'identifiant ARK restera le même et redirigera toujours vers la même ressource. Même dans 50 ans ... en théorie !

ℹ️ Dans le domaine du Web sémantique, l'action d'utiliser un URI pour obtenir la ressource qu'il désigne est appelée "déréférencement d'URI". 

### catalogue.bnf.fr : le coeur bibliographique de la BnF

> 🎬 Cliquez sur cet URI pour accéder à la page publique du catalogue de la BnF affichant la **notice bibliographique** de ... quoi exactement ?

Déjà, on peut remarquer que la ressource qui s'affiche contient la plupart des métadonnées qu'on trouvait dans les ` 🛈 informations détaillées` de la photo numérisée sur Gallica. Pourtant ce n'est pas la même entité : les identifiants ARK sont différents 🧐
On voit sur la droite de la fenêtre un panneau intitulé `Localiser ce document` qui référence 9 exemplaires de la photographie (+ 2 obsolètes) : 3 exemplaires numériques - ce sont ceux sur Gallica -, 3 exemplaires physiques, et 3 exemplaires sur microfilms.

Les notices bibliographiques du catalogue ne décrivent donc pas directement un document concret, mais décrivent une **oeuvre intellectuelle abstraite**  dont les documents conservés sont l'incarnation physique (ou numérique).

En effet, le modèle de métadonnées de la BnF implémente partiellement le modèle conceptuel **WEMI**, issu du domaine des bibliothèques, qui décompose la description d'une oeuvre en quatre niveaux :
```mermaid
graph TD
    %% Définition des classes pour le style
    classDef work fill:#f9f,stroke:#333,stroke-width:2px;
    classDef expression fill:#bbf,stroke:#333,stroke-width:2px;
    classDef manifestation fill:#dfd,stroke:#333,stroke-width:2px;
    classDef item fill:#fff,stroke:#333,stroke-dasharray: 5 5;

    W["<b>1. Œuvre (Work)</b><br/>Un création intellectuelle abstraite.<br/><i>Ex. Harry Potter and the<br/>Philosopher's Stone</i>"]:::work
    
    E["<b>2. Expression</b><br/>La réalisation de l'œuvre dans une forme précise (langue, édition, média)<br/><i>Ex : traduction française *Harry Potter à l'école des sorciers* de Jean-François Ménard</i>"]:::expression
    
    M["<b>3. Manifestation</b><br/>Produit éditorial<br/><i>Ex: Édition Gallimard Jeunesse, 1998<br/>Notice BnF : https://catalogue.bnf.fr/ark:/12148/cb37032584p</i>"]:::manifestation
    
    I["<b>4. Item (Exemplaire)</b><br/>Reproduction physique/numérique<br/><i>Cote : 8-CNLJF-39901<br/>(Tolbiac - Rez-de-jardin)</i>"]:::item

    %% Relations
    E -->|réalise| W
    M -->|matérialise| E
    I -->|exemplifie| M

    %% Note sur la notice
    note1("<b>Ce que décrit la<br/>notice bibliographique</b>")
    note1 -.-> M
```

📚 Plus d'informations sur WEMI/OEMI : https://fr.wikipedia.org/wiki/Mod%C3%A8le_de_r%C3%A9f%C3%A9rence_IFLA_pour_les_biblioth%C3%A8ques


Voici le schéma équivalent pour notre photographie de Eugène Atget :

```mermaid
graph TD
    %% Styles
    classDef work fill:#f9f,stroke:#333,stroke-width:2px;
    classDef expression fill:#bbf,stroke:#333,stroke-width:2px;
    classDef manifestation fill:#dfd,stroke:#333,stroke-width:2px;
    classDef phys fill:#cdff9c,stroke:#333,stroke-width:1px;
    classDef num fill:#9cacff,stroke:#01579b,stroke-width:1px;
    classDef micro fill:#ffc79c,stroke:#455a64,stroke-style:italic;

    W["<b>1. Œuvre</b><br/>Vision artistique d'Atget : Cabaret du Soleil d'or"]:::work
    E["<b>2. Expression</b><br/>Capture photographique (1899)<br/>Négatif sur verre original"]:::expression
    M["<b>3. Manifestation</b><br/>Notice ARK: cb402758340"]:::manifestation

    subgraph Items_Physiques ["4c. Items Physiques (Originaux)"]
        P1["Tirage Ex. 1<br/>(EO-109-3)"]:::phys
        P2["Tirage Ex. 2<br/>(EO-109-3)"]:::phys
        P3["Tirage Réserve<br/>(RES PHOTO OA-173)"]:::phys
    end

    subgraph Items_Numeriques ["4b. Items Numériques (Gallica)"]
        N1["IFN-10506998<br/>ARK@Gallica : btv1b10506998t"]:::num
        N2["IFN-10516963<br/>ARK@Gallica : btv1b10516963n"]:::num
        N3["IFN-10516961<br/>ARK@Gallica : btv1b10516961r"]:::num
    end

    subgraph Items_Substituts ["4a. Items Microformes"]
        S1["MFILM H-25509"]:::micro
        S2["MFILM T-39590"]:::micro
        S3["MFILM G-45733"]:::micro
    end

    %% Relations
    W --> E
    E --> M
    M --> P1 & P2 & P3
    M --> N1 & N2 & N3
    M --> S1 & S2 & S3
```

Le catalogue de la BnF est donc au cœur de son système d'information exposé au public, c'est le pivot qui relie **l'univers bibliographique** (les notices bibliographiques), **l'univers des autorités** (thésaurus, notices d'auteurs) et **l'univers numérique** (Gallica). Il fédère et expose toutes les **métadonnées descriptives** des documents conservés par la bibliothèque, dans des formats lisibles par les humains.

La BnF envisage aujourd'hui son catalogue non pus seulement comme un index explorable par des humains, mais comme un **réservoir de données** exploitable par des machines.
Dans ce but, le catalogue a été transformé en un très grand **graphe de connaissances** exposé publiquement sur [data.bnf.fr](data.bnf.fr]).

### data.bnf.fr : le catalogue en graphe requêtable

> 🎬 Rendez-vous sur [data.bnf.fr](data.bnf.fr) et recherchez `Eugène Atget`. Vous arrivez sur la page d'autorité qui agrège toutes les ressources du graphe de connaissances liées à cet auteur. Vérifiez que parmi ses oeuvres se trouvent est bien recensée la photographie du  `cabaret du soleil d'or`.

ℹ️ il faut cocher `☑ Éditions` pour afficher toutes les **Manifestations** associées aux oeuvres de l'auteur.

Vous venez d'expérimenter l'interface *human-friendly* d'accès au graphe, mais la véritable puissance de data.bnf.fr est ailleurs : c'est un point d'accès machine. Son but est de permettre **l'interrogation automatisée** du graphe de connaissances de la BnF.

Les données y sont structurées selon le modèle RDF (Resource Description Framework), sous forme de triplets sémantiques. Ces triplets forment un réseau complexe (le graphe de connaissances) que l'on peut interroger via le langage de requête SPARQL.

ℹ️ Rappel : Un triplet est un tuple (sujet, propriété, objet) qui exprime une affirmation simple et qui correspond dans le graphe à un triplet `[noeud] -- (arête dirigée étiquetée) --> [noeud]`. Par exemple : `[<Eugène Atget>]--(est auteur de)-->(photographie du cabaret du soleil d'or)`. Le sujet est toujours une ressource, identifiée par un URI. L'objet peut être aussi une ressource (URI), ou bien un *literal* : un nombre, une chaîne de caractère, etc. La propriété (ou prédicat) définit la relation. Elle possède également un URI défini dans une ontologie et peut être désignée par un préfixe, par exemple `rdf:type` pour définir la nature d'une ressource.

> 🎬Le site [data.bnf.fr](data.bnf.fr) offre un point d'accès SPARQL graphique destiné aux humains assez motivés pour tenter d'accéder directement au graphe.
> Accédez-y depuis le menu `Données → Sparql`.
> Commençons par le plus simple : obtenir tous les triplets associés à une ressource. L'interface propose des exemples de requêtes dans un menu déroulant situé en haut à gauche de la fenêtre.
> 1. Choisissez `Obtenir tous les triplets contenant une ressource` pour charger la requête préparée. 
> 2. Éditez l'URI de la ressource chargée par défaut par celle de notre photographie (on rapelle que son id ARK est `cb402758340`), puis exécutez la requête pour récupérer les triplets associés.

Voici, représenté sous forme de graphe, ce que vous devriez récupérer :

```mermaid
graph TD
    classDef doc fill:#ffabab,stroke:#333,stroke-width:2px;
    classDef expression fill:#bbf,stroke:#333,stroke-width:2px;
    classDef manifestation fill:#dfd,stroke:#333,stroke-width:2px;
    
    %% Définition des nœuds (Sujet et Objets URIs)
    S["http://data.bnf.fr/ark:/12148/cb402758340"]:::doc
    Type["http://www.w3.org/2004/02/skos/core#Concept"]
    About["http://data.bnf.fr/ark:/12148/cb402758340#about"]:::manifestation

    %% Définition des Littéraux (Data Properties)
    Created(["1986-10-13"])
    Modified(["2022-12-13"])

    %% Arêtes (Prédicats)
    S -- "rdf:type" --> Type
    S -- "dcterms:created" --> Created
    S -- "dcterms:modified" --> Modified
    S -- "foaf:focus" --> About

    %% Style pour distinguer URIs et Littéraux
    style Type fill:#e1f5fe,stroke:#01579b
    style Created fill:#fff,stroke:#333,stroke-dasharray: 5 5
    style Modified fill:#fff,stroke:#333,stroke-dasharray: 5 5
```

Un peu...décevant, non ? Où sont toutes les métadonnées ? 🫤

Dans le graphe de data.bnf.fr, les URI "racines" comme `http://data.bnf.fr/ark:/12148/cb402758340` désignent la notice bibliographique elle-même et fonctionnent comme des points d'entrée dans le graphe. En fait, le graphe ne représente pas explicitement l'entité **Work/Oeuvre** du modèle WEMI qui est remplacée par ces "noeuds-notices".

C'est à partir de ce point qu'on peut accéder aux resources décrivant l'oeuvre conservée lui-même. C'est à dire selon le modèle WEMI à l'**expression** de cette oeuvre et mieux encore sa **manifestation**.

> 🎬 En fait, le graphe récupéré nous dit que le point d'entrée référence directement les manifestations de l'oeuvre grâce à la propriété `foaf:focus`. 
> Donc, `http://data.bnf.fr/ark:/12148/cb402758340#about` est l'URI de la **la manifestation** de cette oeuvre, c'est à dire sa **représentation matérielle** ! Modifiez la requête pour décrire cette ressource.

ℹ️ La BnF utilise élégamment les [**fragments d'URI**](https://en.wikipedia.org/wiki/URI_fragment) - le suffix commençant par # - pour signifier que les manifestations & expressions sont des ressources secondaires dérivées d'une ressource primaire : la notice.

Vous obtiendrez les triplets décrivant ce graphe, ici simplifié :

```mermaid
graph TD
    %% Styles
    classDef work fill:#f9f,stroke:#333,stroke-width:2px;
    classDef expression fill:#bbf,stroke:#333,stroke-width:2px;
    classDef manifestation fill:#dfd,stroke:#333,stroke-width:2px;
    classDef type fill:#e1f5fe,stroke:#01579b;
    classDef literal fill:#fff,stroke:#333,stroke-dasharray: 5 5;
    classDef resource fill:#fff9c4,stroke:#455a64;

    %% Nœud pivot
    About(".../cb402758340#about"):::manifestation

    %% Identité et Type
    About -- "foaf:focus" --- Notice("Notice ARK"):::resource
    About -- "rdf:type" --> T1("Manifestation (RDA/FRBR)"):::type

    %% Bloc Description (Regroupement des littéraux)
    subgraph Description ["Données descriptives"]
        About -- "dc:title / date" --> Title["'Cabaret du Soleil d'or' (1899)"]:::literal
        About -- "dc:description / note" --> Tech["Technique : Papier albuminé, bromure..."]:::literal
        About -- "bnf:FRBNF" --> ID["'40275834'"]:::literal
    end

    %% Bloc Accès (Regroupement des liens sortants)
    subgraph Acces_et_Indexation ["Ressources liées"]
        About -- "electronicReproduction (x3)" --> Gallica["Items Gallica (6998t, 6961r, 6963n)"]:::resource
        About -- "dc:subject (x6)" --> Rameau["Concepts Sujets (RAMEAU)"]:::resource
    end

    %% Lien structurel
    About -- "expressionManifested" --> Expr["#Expression"]:::resource
```

On voit bien qu'il s'agit d'une ressource de type **manifestation** qui représente notre photo du Cabaret du Soleil d'Or, qui possède une partie des métadonnées de la photographie comme son titre, sa date ou sa description, ainsi que des sujets liés pris dans le thésaurus RAMEAU de la BnF et tous référencés sur data.bnf.fr.
La ressource pointe aussi vers les trois numérisations de cette photographie qui sont, au sens de WEMI, trois **items**.
Enfin, cette manifestation pointe vers **l'expression** qu'elle matérialise. Son URI est formée de celle de l'oeuvre, suffixée de `#Expression`

> 🎬 Reprenez la requête pour décrire cette fois **l'expression** de l'oeuvre, c'est à dire son contenu **intellectuel et artistique**.

Vous obtiendrez l'extrait de graphe ci-dessous.

ℹ️ Remarquez les items physiques (microfilms, exemplaires papier) ne sont pas référencés, mais seulement les numérisations sur Gallica.

```mermaid
graph TD
    %% Styles
    classDef work fill:#f9f,stroke:#333,stroke-width:2px;
    classDef expression fill:#bbf,stroke:#333,stroke-width:2px;
    classDef manifestation fill:#dfd,stroke:#333,stroke-width:2px;
    classDef type fill:#e1f5fe,stroke:#01579b;
    classDef literal fill:#fff,stroke:#333,stroke-dasharray: 5 5;
    classDef resource fill:#fff9c4,stroke:#455a64;

    %% Nœuds de transition (depuis la Manifestation)
    About(".../cb402758340#about"):::manifestation
    Expr(".../cb402758340#Expression"):::expression

    %% Liens structurels WEMI
    About -- "rdam:P30139<br/>expressionManifested" --> Expr

    %% Catégorie : Nature et Format (Types)
    subgraph Nature_du_Contenu ["Définition de l'Expression"]
        Expr -- "rdf:type" --> T1("rda:Expression"):::type
        Expr -- "rdf:type" --> T2("rdac:C10006"):::type
        Expr -- "dc:type" --> T3("StillImage"):::type
        Expr -- "dc:language" --> Lang["ISO: fre"]:::literal
    end

    %% Catégorie : Responsabilités (Contributeurs et Rôles)
    subgraph Responsabilites ["Agents et Rôles"]
        Atget("Eugène Atget<br/>(...cb11889340k#about)"):::resource
        Jammes("Marie-Thérèse Jammes<br/>(...cb12541292d#about)"):::resource

        Expr -- "Atget :<br/>contributor<br/>relators:pht<br/>bnf-role:r530" --> Atget
        Expr -- "Jammes :<br/>contributor<br/>relators:col<br/>bnf-role:r4030" --> Jammes
    end

    %% Catégorie : Indexation (Sujets RAMEAU)
    subgraph Indexation_Sujets ["Thématiques"]
        Expr -- "dc:subject" --> S1["Sujet 1"]:::resource
        Expr -- "dc:subject" --> S2["Sujet 2"]:::resource
        Expr -- "dc:subject" --> S3["Sujet 3..."]:::resource
    end
```

On constate que **l'expression** est utilisée pour lier des ressources issues des référentiels d'autorité de la BnF : notices d'auteurs, sujets issus du thésaurus RAMEAU.

La photographie du `Cabaret du Soleil d'Or`, dont la fiche bibliographique est [catalogue.bnf.fr/ark:/12148/cb402758340](catalogue.bnf.fr/ark:/12148/cb4027583400), est donc représentée dans le graphe [data.bnf.fr](data.bnf.fr) par trois entités principales : le **point d'entrée "notice"**, **l'expression** et la **manifestation** de l'oeuvre. Les numérisations sont liées à la manifestation comme **items** du modèle WEMI.

Résumé grahiquement :

```mermaid
graph TD
    %% Styles
    classDef doc fill:#ffabab,stroke:#333,stroke-width:2px;
    classDef expression fill:#bbf,stroke:#333,stroke-width:2px;
    classDef manifestation fill:#dfd,stroke:#333,stroke-width:2px;
    classDef item fill:#fff,stroke:#333,stroke-dasharray: 2 2;
    classDef uri fill:#fff9c4,stroke:#455a64,font-family:monospace,font-size:10px;

    %% ENTITÉS (URI)
    D["<b>1. Point d'entrée (Notice)</b><br/>URI : http://data.bnf.fr/ark:/12148/cb402758340"]:::doc
    E["<b>2. Expression</b><br/>Informations d'autorités sur l'oeuvre<br/>URI : http://data.bnf.fr/ark:/12148/cb402758340#Expression"]:::expression
    M["<b>3. Manifestation</b><br/>Informations sur le document physique conservé<br/>URI : http://data.bnf.fr/ark:/12148/cb402758340#about"]:::manifestation

    subgraph Exemplaires ["4. Items (Exemplaires)"]
        N1[".../btv1b10506998t"]:::item
        N2[".../btv1b10516963n"]:::item
        N3[".../btv1b10516961r"]:::item
    end

    %% RELATIONS (PRÉDICATS)
    
    %% Doc -> Expression
    D -- "foaf:focus<br/>" --> M
    
    %% Expression -> Manifestation
    M -- "rdam:P30139<br/>(expressionManifested)" --> E
    
    %% Manifestation -> Items
    M -- "rdam:P30011<br/>(electronicReproduction)" --> N1
    M -- "rdam:P30011<br/>(electronicReproduction)" --> N2
    M -- "rdam:P30011<br/>(electronicReproduction)" --> N3
```
